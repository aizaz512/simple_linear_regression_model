```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel('/xcxcxc.xlsx')
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Panel_Area_m^2'], df['Energy_kWh'])
plt.xlabel('Panel Area (m^2)', color = "red")
plt.ylabel('Energy (kWh)', color = "red")
plt.title('Panel Area vs. Energy')
plt.show()
```

```python
lr = linear_model.LinearRegression()
lr.fit(df[['Panel_Area_m^2']], df.Energy_kWh)
```

```python
lr.predict([[34]])
```

```python
lr.score(df[['Panel_Area_m^2']], df.Energy_kWh)
lr.coef_
```

```python
lr.intercept_
```

```python
predicted = 0.76074117 * 34 + 0.019918934568613267
print(predicted)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Panel_Area_m^2'], df['Energy_kWh'])
plt.xlabel('Panel Area (m^2)', color = "red")
plt.ylabel('Energy (kWh)', color = "red")
plt.title('Panel Area vs. Energy')
plt.plot(df['Panel_Area_m^2'], lr.predict(df[['Panel_Area_m^2']]), color = 'blue')
plt.show()
```
